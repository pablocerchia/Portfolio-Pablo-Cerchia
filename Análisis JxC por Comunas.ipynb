```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from functools import reduce
```

Cuestiones preeliminares antes de comenzar con el informe de campaña.

```python
elecciones_raw = pd.read_csv('https://raw.githubusercontent.com/MET4OP-UBA/MET4OP/master/dataset/elecciones_2019/resultados/120819-054029/datos_agrup.csv')
```

```python
#Seleccionamos las columnas que queremos usar
elecciones1 = elecciones_raw[["VOTOS_AGRUPACION", "NOMBRE_CATEGORIA", "NOMBRE_REGION", "NOMBRE_AGRUPACION"]]
#Filtramos filas en la columna NOMBRE_CATEGORIA
elecciones1[elecciones1["NOMBRE_CATEGORIA"]== 'Presidente y Vicepresidente de la República']
e_presidenciales = elecciones1[elecciones1["NOMBRE_CATEGORIA"]== 'Presidente y Vicepresidente de la República']
```

```python
#Filtramos el dataframe e_presidenciales por filas (Comunas) dentro de la columna NOMBRE_REGION  
#Lo hacemos para cada una de las comunas 
#COMUNA 1
Comuna1 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 01']
#Agrupamos en cada una de las comunas a las agrupaciones/partidos y sumamos los votos que tienen cada uno
#Armamos un nuevo dataframe por comuna con los datos de votos totales por partido
com1 = Comuna1.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 2 
Comuna2 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 02']

com2 = Comuna2.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 3
Comuna3 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 03']

com3 = Comuna3.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 4
Comuna4 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 04']

com4 = Comuna4.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 5
Comuna5 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 05']

com5 = Comuna5.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 6
Comuna6 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 06']

com6 = Comuna6.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 7
Comuna7 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 07']

com7 = Comuna7.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 8
Comuna8 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 08']

com8 = Comuna8.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 9
Comuna9 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 09']

com9 = Comuna9.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 10
Comuna10 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 10']

com10 = Comuna10.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 11
Comuna11 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 11']

com11 = Comuna11.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 12
Comuna12 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 12']

com12 = Comuna12.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 13
Comuna13 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 13']

com13 = Comuna13.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 14
Comuna14 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 14']

com14 = Comuna14.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()

#COMUNA 15
Comuna15 = e_presidenciales[e_presidenciales["NOMBRE_REGION"]=='COMUNA 15']

com15 = Comuna15.groupby(["NOMBRE_AGRUPACION"])["VOTOS_AGRUPACION"].sum()
```

```python
#Recopilamos los dataframes para mergear
dataframes = [com1, com2, com3, com4, com5, com6, com7, com8, com9, com10, com11, com12, com13, com14, com15]
tabla_comunas = reduce(lambda  left,right: pd.merge(left,right,on=['NOMBRE_AGRUPACION'],
                                            how='outer'), dataframes)
#Cambiamos los nombres de la columnas
tabla_comunas.columns= ["Comuna1", "Comuna2","Comuna3","Comuna4","Comuna5","Comuna6","Comuna7","Comuna8","Comuna9","Comuna10","Comuna11","Comuna12","Comuna13","Comuna14","Comuna15"] 

#Quitamos de la tabla algunos tipos de voto que son irrelevantes para calcular el porcentaje. La explicación de por que dejamos el voto en blanco en el cálculo está en el informe.
tablapresidencial=tabla_comunas.drop(["RECURRIDO","IMPUGNADO", "NULO"])

tablapresidencial
```

```python
#Calculamos el porcentaje de la comuna que le digamos y agrega una columna al final para utilizar luego en el informe.
tablapresidencial['% Com1'] = (tablapresidencial['Comuna1'] / 
                    tablapresidencial['Comuna1'].sum()) * 100 
tablapresidencial['% Com2'] = (tablapresidencial['Comuna2'] / 
                    tablapresidencial['Comuna2'].sum()) * 100 
tablapresidencial['% Com3'] = (tablapresidencial['Comuna3'] / 
                    tablapresidencial['Comuna3'].sum()) * 100 
tablapresidencial['% Com4'] = (tablapresidencial['Comuna4'] / 
                    tablapresidencial['Comuna4'].sum()) * 100 
tablapresidencial['% Com5'] = (tablapresidencial['Comuna5'] / 
                    tablapresidencial['Comuna5'].sum()) * 100 
tablapresidencial['% Com6'] = (tablapresidencial['Comuna6'] / 
                    tablapresidencial['Comuna6'].sum()) * 100 
tablapresidencial['% Com7'] = (tablapresidencial['Comuna7'] / 
                    tablapresidencial['Comuna7'].sum()) * 100 
tablapresidencial['% Com8'] = (tablapresidencial['Comuna8'] / 
                    tablapresidencial['Comuna8'].sum()) * 100 
tablapresidencial['% Com9'] = (tablapresidencial['Comuna9'] / 
                    tablapresidencial['Comuna9'].sum()) * 100 
tablapresidencial['% Com10'] = (tablapresidencial['Comuna10'] / 
                    tablapresidencial['Comuna10'].sum()) * 100 
tablapresidencial['% Com11'] = (tablapresidencial['Comuna11'] / 
                    tablapresidencial['Comuna11'].sum()) * 100 
tablapresidencial['% Com12'] = (tablapresidencial['Comuna12'] / 
                    tablapresidencial['Comuna12'].sum()) * 100 
tablapresidencial['% Com13'] = (tablapresidencial['Comuna13'] / 
                    tablapresidencial['Comuna13'].sum()) * 100 
tablapresidencial['% Com14'] = (tablapresidencial['Comuna14'] / 
                    tablapresidencial['Comuna14'].sum()) * 100 
tablapresidencial['% Com15'] = (tablapresidencial['Comuna15'] / 
                    tablapresidencial['Comuna15'].sum()) * 100
```

Análisis general

El resultado de las PASO 2019 para Presidente en la ciudad de Buenos Aires ha sido muy favorable para nuestro partido (Juntos por el Cambio). En un plano general y en un análisis de los resultados totales, encontramos que JxC aventaja al Frente de Todos en casi 12 puntos: 44,71% contra 33.04% respectivamente como se muestran en los gráficos 1 y 2 y las tablas a continuación. En ellas también se observa que JxC se impuso en 12 de las 15 comunas. En las restantes tres el frente de TODOS salió victorioso.

Este cálculo ha sido hecho incluyendo al % correspondiente al voto en blanco. Si bien metodológicamente sabemos que se debería retirar esta categoría del análisis para calcular el porcentaje exacto, consideramos que el voto en blanco debe ser observado y analizado de cara a las elecciones generales. Creemos que una adecuada estrategia de polarización generaría que esos votos en blanco se traduzcan en votos hacia nuestra fuerza.

Sabemos que el principal adversario de JxC es el frente de TODOS. Según nuestras encuestas a lo largo de los últimos años conocemos que la última fuerza tiene un núcleo duro inamovible de alrededor del 30% del padrón de CABA. Por nuestra parte, el núcleo duro de JxC en CABA es de alrededor del 40%. Estos datos son confirmados por la columna de promedio en la tabla que está a continuación. 
Sin embargo, consideramos que, en la Ciudad, nuestro potencial de crecimiento es muy superior al del frente de TODOS. Por ello creemos adecuado hacer énfasis en una estrategia de campaña dura de polarización y sobre la inutilidad del voto destinado a terceras fuerzas. El principal objetivo será entonces obtener aquellos votos conseguidos por el partido Consenso Federal pero también otros partidos menores como el Frente NOS y Unite por la Libertad.

Nuestra investigación arrojó que a entre un 6% y 7% del padrón electoral lo podríamos considerar votantes “duros” de Consenso Federal. Dichos votantes demandarían muchos recursos para ser convencidos y de todas formas la probabilidad de éxito es bajo. Sin embargo, hemos observado que en varias comunas, Consenso Federal estuvo por encima de su promedio nacional e incluso su promedio en las 15 comunas está en 8.73%. En algunas incluso ha superado el 10% de los sufragios. Una campaña de polarización tendrá entonces que estar destinada especialmente a ese 3 o 4% de electores que se han inclinado por Consenso Federal y que forman parte de los votantes “blandos” o volátiles cuyos votos pueden variar. Una estrategia similar de polarización servirá para captar votos de las otras dos fuerzas mencionadas con las cuales JxC posee, con respecto a determinados temas, cercanía ideológica. La importancia del voto útil en la campaña es central para obtener dichos votos que podrían aportar entre un 1 y 2% extra a nuestra fuerza.




```python
#Imprimimos los resultados totales a nivel caba
tablapresidencial["Totales"] = tablapresidencial.sum(axis=1)
tablapresidencial["%Totales"] = ((tablapresidencial['Totales'] / 
                    tablapresidencial['Totales'].sum()) * 100).round(2) 

tablapresidencial[["%Totales"]]
```

```python
#Armamos una tabla sola con porcentajes
porcentajes_presidencial = tablapresidencial.drop(["Comuna1", "Comuna2","Comuna3","Comuna4","Comuna5","Comuna6","Comuna7","Comuna8","Comuna9","Comuna10","Comuna11","Comuna12","Comuna13","Comuna14","Comuna15"], axis=1)
porcentajes_presidencial = porcentajes_presidencial.round(2)
#Armamos categoria residual en donde incluimos a las filas que no nos interesan. Acá se agrupan las fuerzas a las cuales no es posible quitarle votos por una distancia ideológica muy grande (principalmente el FIT y el MAS) o aquellas fuerzas que no pasaron las PASO (Frente Patriota, Movimiento de Accion vecinal y partido autonomista)
residual = porcentajes_presidencial.iloc[[2, 5, 7, 8, 9]]
print(residual.sum())
```

```python
#Eliminamos las filas que fueron agrupadas en la categoría residual
porcentajes_presidencial_final = porcentajes_presidencial.drop(["FRENTE DE IZQUIERDA Y DE TRABAJADORES - UNIDAD", "FRENTE PATRIOTA", "MOVIMIENTO AL SOCIALISMO", "MOVIMIENTO DE ACCION VECINAL", "PARTIDO AUTONOMISTA"])
porcentajes_finales=porcentajes_presidencial_final.drop(["Totales","%Totales"], axis=1)
#Agregamos fila "RESIDUAL" a la tabla porcentajes_presidencial
porcentajes_finales.loc["RESIDUAL"]=[5.15, 3.65, 6.25, 6.01, 6.67, 5.90, 5.72, 5.32, 5.59, 6.18, 5.69, 5.75, 4.55, 4.28, 6.58]
#Eliminamos las filas que componen a categoría "RESIDUAL"
porcentajes_finales["Promedio"]= porcentajes_finales.mean(axis=1).round(2)

porcentajes_finales
```
